## **Importing required Packages**


```python
import numpy as np
import pandas as pd
```

# **Reading csv file**


```python
car_detail_csv = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ml collections/car_specifications_csv.txt")
car_detail_csv
```

# **Reading Excel File**


```python
car_price = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/ml collections/car_pricing_loan_amount.xlsx")
car_price
```

# **Converting into DataFrame**


```python
pd.DataFrame(car_detail_csv)
```

```python
pd.DataFrame(car_price)
```

# **Merging two DataFrames**

```python
merge_data = pd.concat([car_detail_csv,car_price])
merge_data
```

# **Filling NA values**

```python
data = merge_data.fillna(0)
data
```

# **Iterate over to add gst value in on road pricing**
