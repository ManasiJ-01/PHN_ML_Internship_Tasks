# **Q.4) Finding Statistics Aggregate**

```python
df = data[['Age','Fare']]
age_fare = pd.DataFrame(df)
stats_agg = age_fare.agg(['sum','min','max','count','mean','std','median','skew'])
stats_agg
```

```python
# OR

data.agg(
    {
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
```

# **Q.5) Finding mean ticket fare price for each of the sex and cabin class combinations**

```python
data.groupby(['Sex','Pclass'])['Fare'].mean()
```
