```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
categories = np.array(['A','B','C','D','E'])
values = np.array([23,56,41,62,19])
plt.subplot(2,2,1)
plt.bar(categories,values)
plt.xlabel('Category')
plt.ylabel('Values')

data = np.array([12,17,21,18,14,13,16,9,12,15,19,11,14,16,20,18,15,13,16,11,19])
plt.subplot(2,2,2)
plt.hist(data)


labels = np.array(['Apple','Banana','Orange','Mango'])
size = np.array([30,25,15,30])
plt.subplot(2,2,3)
plt.pie(size)

x_values = np.array([1,2,3,4,5])
y_values = np.array([2,5,3,6,4])
plt.subplot(2,2,4)
plt.scatter(x_values,y_values)

plt.show()
```
